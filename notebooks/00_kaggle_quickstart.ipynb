# CircuitLens — Kaggle Quickstart

This notebook downloads a small causal language model, captures reduced attention and MLP activation data, and writes a `.llmgraph.zip` file to `/kaggle/working`.

```python
!git clone https://github.com/shaig-mahmudov/circuit-lens.git
%cd circuit-lens
!pip install -q --no-build-isolation -e ./packages/engine
```

```python
from circuit_lens import AnalysisConfig, analyze, export_bundle

MODEL_ID = 'gpt2'
PROMPT = 'Baku is the capital of'
TARGET_TOKEN = ' Azerbaijan'

config = AnalysisConfig(
    model_id=MODEL_ID,
    prompt=PROMPT,
    target_token=TARGET_TOKEN,
    device='auto',
    dtype='auto',
    top_logits=10,
    top_neurons=30,
    top_attention_edges=5,
    minimum_attention=0.02,
)
```

```python
bundle = analyze(config)
output = export_bundle(bundle, '/kaggle/working/gpt2-baku.llmgraph.zip')
print(output)
```

```python
print('Model:', bundle.model.id)
print('Tokens:', [token.display for token in bundle.tokens])
print('Attention graphs:', len(bundle.attention.graphs))
print('Neuron nodes:', len(bundle.neurons.nodes))
print('Top predictions:')
for item in bundle.logits.top:
    print(f'{item.token!r}: {item.probability:.2%}')
```
